```python
pmtdf = DB.DataPMT()
sipmdf = DB.DataSiPM()

adc_to_pes = -1.0/pmtdf["adc_to_pes"].reshape(12,1)
```

```python
def events(print_mod=100):
    i = 0
    for filename in sorted(glob.glob("/Users/Gonzalo/github/IC/data/DATA/run_3103/*.h5")):
        with tb.open_file(filename) as h5f:
            for j, evt in enumerate(h5f.root.RD.pmtcwf):
                t = h5f.root.Run.events.timestamp[j]
                yield i, t, evt
                if not i%print_mod:
                    print("event",i)
                i += 1
```

```python
t = []
```

```python
twf = np.arange(48000)/40.

def sum_wf(wfs):
    return np.sum(wfs * adc_to_pes, axis=0)

def zs_wf(wf, thrs=0.5):
    wf = sum_wf(wf)
    df = wfm.wf2df(twf, wf)
    return wfm.wf_thr(df, thrs)

def find_peaks(wf, stride=100., min_length=100.):
    peaks = []
    t, e = wf.time_mus.values, wf.ene_pes.values
    begin = 0
    stride /= 25.
    min_length = int(min_length/25.)
    for i in range(1, t.size):
        if t[i] - t[i-1] > stride:
            if i - begin > min_length:
                peaks.append(wfm.wf2df(t[begin:i], e[begin:i]))
            begin = i
    if begin != i:
        peaks.append(wfm.wf2df(t[begin:i], e[begin:i]))
    return peaks

def plot_peaks(peaks):
    fig = plt.figure()
    fig.set_size_inches(9,7)
    for peak in peaks:
        fig.clear()
        plt.plot(peak.time_mus.values, peak.ene_pes.values)
        plt.show()
        
        raw_input("size: {} mus; E: {:.2} pes; time: {}->".format(len(peak)/40.,
                                                                  peak.ene_pes.sum(),
                                                                  peak.time_mus[np.argmax(peak.ene_pes.values)]))
```

```python
XRs = filter(lambda pmap: len(filter(isS1, pmap))==1 and len(filter(isXray, pmap))==1, pmaps)
```

```python
z, e = [], []
for pmap in XRs:
    foundS1 = False
    for peak in pmap:
        if isS1(peak):
            foundS1 = True
            t0 = peak.time_mus.values[np.argmax(peak.ene_pes.values)]
        elif isXray(peak) and foundS1:
            z.append(peak.time_mus.values[np.argmax(peak.ene_pes.values)] - t0)
            e.append(peak.ene_pes.values.sum())
z, e = map(np.array, [z, e])
e /= 12.
plt.scatter(z, e, marker=".", label="data")
plt.xlim(0, 600)
plt.ylim(0, 8e2)#plt.ylim()[1])
x, y, yerr = fit.profileX(z, e, 60, (0, 600))
plt.plot(x, y, "g--", label="profile", linewidth=3)
f, v, e = fit.fit("expo", x, y, (4500., -1e4))
x = np.linspace(0, 600, 1000)
plt.plot(x, f(x), "r-", label="fit", linewidth=2)
amp, slope = v
plt.text(100, 600, "Lifetime = {:.2f} ms".format(-slope/1000))
plt.xlabel("Drift time ($\mu$s)")
plt.ylabel("Energy (pes)")
print(v)
```

```python
isS2 = lambda x: not isS1(x)
s1, s2 = [], []
for pmap in pmaps:
    if not len(filter(isS1, pmap))==1:
        continue
    
    s1s = filter(isS1, pmap)[0]
    t0 = s1s.time_mus.values[np.argmax(s1s.ene_pes.values)]
    s1.append(s1s.ene_pes.values.sum())
    
    s2s = filter(isS2, pmap)
    if len(s2s) == 0:
        s1.pop()
        continue
    enes = [peak.ene_pes.values.sum() for peak in s2s]
    maxpeak = np.argmax(enes)
    if s2s[maxpeak].time_mus.values[0] < t0:
        s1.pop()
        continue
    s2.append(max(enes))

s1, s2 = map(np.array, [s1, s2])
plt.scatter(s1, s2, marker=".", label="data")
plt.ylim(0,3e5)
x, y, yerr = fit.profileX(s1, s2, 20, (200, 450))
plt.plot(x, y, "g--", label="profile", linewidth=3)
f, v, e = fit.fit("polynom", x, y, (0., 1e4))
plt.plot(x, f(x), "r-", label="fit", linewidth=2)
print(v)
plt.xlabel("S1 energy (pes)")
plt.ylabel("S2 energy (pes)")

plt.figure()
h = plt.hist(s2/s1, 100, range=(0, 1000))
x, y = h[1][:-1], h[0]
x += np.diff(h[1])/2.
f, v, e = fit.fit("gauss", x, y, (1e4, 500., 50.))
x = np.linspace(*plt.xlim(), num=500)
plt.plot(x, f(x), "r-", label="fit", linewidth=2)
print(v)
amp, mu, sigma = v
plt.text(800, 80, "$\mu$ = {:.1f}\n$\sigma$ = {:.2f}".format(mu, sigma))
plt.xlabel("S2 / S1 ratio")
plt.ylabel("Entries")
```
