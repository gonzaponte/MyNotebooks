```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from functools import lru_cache
```

```python
def define(n):
    @lru_cache(maxsize=n+1)
    def f(k):
        if k <= 0: return 0
        if k == 1: return 1
        return sum(f(k-1-i) for i in range(n))
    
    return f
```

```python
n      = 10
N      = 15
output = np.empty((n, N))

for i in range(10):
    f = define(i+1)
    output[i] = [f(k) for k in range(N)]
```

```python
output.shape
```

```python
plt.figure(figsize=(12, 8))
plt.plot(np.arange(N)+1, output.T)
#plt.xscale("log")
plt.yscale("log")
```
