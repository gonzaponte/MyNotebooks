```python
import numpy as np
import matplotlib.pyplot as plt

from CCplot import auto_plot_style
auto_plot_style()
```

```python
x  = np.logspace(-5, 0, 200), np.linspace(1, 20, 1001)
x  = np.sort(np.concatenate(x))
y1 = 1   * np.exp(-x/20)
y2 = 1/2 * np.exp(-x/20) + 1/2 * np.exp(-x*10)
y3 = 1/3 * np.exp(-x/20) + 1/3 * np.exp(-x*10) + 1/3 * np.exp(-x*1000)
```

```python
plt.plot(x, y1, label="1 exp")
plt.plot(x, y2, label="2 exp")
plt.plot(x, y3, label="3 exp")
plt.xscale("log")
plt.yscale("log")
plt.legend()
```
