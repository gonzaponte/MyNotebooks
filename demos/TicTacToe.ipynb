```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
games = []
p = 1
for i1 in range(9):
    for i2 in range(9):
        for i3 in range(9):
            for i4 in range(9):
                for i5 in range(9):
                    for i6 in range(9):
                        for i7 in range(9):
                            for i8 in range(9):
                                for i9 in range(9):
                                    indices = [i1, i2, i3, i4, i5, i6, i7, i8, i9]
                                    if len(set(indices)) < 9: continue
                                
                                    g = np.zeros(9)
                                    g[indices] = [1, -1, 1, -1, 1, -1, 1, -1, 1]
                                    g.reshape(3, 3)
                                    games.append(g)
```

```python
def plot(g):
    bins = [-1.5, -0.5, 0.5, 1.5]
    x, y = np.meshgrid([-1, 0, 1], [-1, 0, 1], indexing="ij")
    plt.hist2d(x.flatten(), y.flatten(), bins=bins, weights=g.flatten())
    plt.xticks([])
    plt.yticks([])
```

```python
for i, g in enumerate(games):
    plt.figure()
    plot(g)
    if i>10: break
```
