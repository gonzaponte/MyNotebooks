```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams["figure.figsize"] = 10, 8
plt.rcParams["font.size"] = 20
```

```python
x  = np.linspace(-1, 1, 1001)
y  = x
xx, yy = map(np.ravel, np.meshgrid(x, y, indexing="ij"))
```

```python
light = np.abs(np.abs(xx) + np.abs(yy) - 1) < 1e-2
zz    = np.zeros(xx.size)
zz[light] = 1
zz    = zz.reshape((x.size, y.size))
plt.imshow(zz)
```

```python
light = np.abs(np.abs(xx)**2 + np.abs(yy)**2 - 1) < 1e-2
zz    = np.zeros(xx.size)
zz[light] = 1
zz    = zz.reshape((x.size, y.size))
plt.imshow(zz)
```

```python
light = np.abs(np.abs(xx)**3 + np.abs(yy)**3 - 1) < 1e-2
zz    = np.zeros(xx.size)
zz[light] = 1
zz    = zz.reshape((x.size, y.size))
plt.imshow(zz)
```

```python
light = np.abs(np.abs(xx)**4 + np.abs(yy)**4 - 1) < 1e-2
zz    = np.zeros(xx.size)
zz[light] = 1
zz    = zz.reshape((x.size, y.size))
plt.imshow(zz)
```

```python
light = np.abs(np.abs(xx)**10 + np.abs(yy)**10 - 1) < 1e-2
zz    = np.zeros(xx.size)
zz[light] = 1
zz    = zz.reshape((x.size, y.size))
plt.imshow(zz)
```

```python
light1 = np.abs(-xx + np.abs(yy) - 1) < 1e-2
light2 = np.abs(xx) < 1e-2
zz    = np.zeros(xx.size)
zz[light1] = 1
zz[light2] = 1
zz    = zz.reshape((x.size, y.size))
plt.imshow(zz)
```
