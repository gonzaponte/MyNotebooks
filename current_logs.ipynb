```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.rcParams["figure.figsize"] = 8, 6
plt.rcParams["font.size"] = 20
plt.rcParams["lines.linewidth"] = 2
```

```python
@np.vectorize
def totime(s):
    h, m, s = map(float, s.split(":"))
    return s + m*60 + h*3600

def load(filename):
    data = pd.read_csv(filename, sep=" ", header=None, names="date time I".split())
    data = data.assign(timestamp = totime(data.time))
    data = data.assign(timestamp = (data.timestamp - data.timestamp.min()) / 60)
    return data
```

```python
filename = "/home/gonzalo/data/LHM/current_logs/chargedown2_20210802.txt"
data = load(filename)

plt.figure(figsize=(10, 6))
plt.plot(data.timestamp, data.I, label="data")
plt.plot(data.timestamp, data.I.rolling(75, center=True).mean(), label="roll ave")
plt.legend()
plt.xlabel("Time (min)")
plt.ylabel("Current (a.u.)")
```
