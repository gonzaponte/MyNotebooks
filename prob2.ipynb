```python
import numpy as np
import matplotlib.pyplot as plt

from collections import defaultdict
from operator    import attrgetter
from operator    import itemgetter
```

```python
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = 12, 8
plt.rcParams["font.size"] = 15
```

```python
def gen(n):
    return np.array(list(map(int, str(2**n))))

def count(n):
    s    = gen(n)
    x, y = np.unique(s, return_counts=True)
    ye   = np.sqrt(y)
    n    = np.sum(y)
    y    = y  / n
    ye   = ye / n
    d    = defaultdict(int)
    du   = defaultdict(lambda *args: np.inf)
    for xi, yi, yei in zip(x, y, ye):
        d [xi] = yi 
        du[xi] = yei
    return d, du
```

```python
for i in map(int, np.random.randint(10, 100000, size=8)):
    s = gen(i)
    x, y = np.unique(s, return_counts=True)
    ye  = np.sqrt(y)
    n   = np.sum(y)
    y   = y  / n
    ye  = ye / n
    plt.errorbar(x, y, ye, fmt="p", label=f"2**{i}")
plt.legend()
```

```python
data, datau = zip(*list(map(count, range(100000))))
```

```python
y = list(map(itemgetter(0), data ))
u = list(map(itemgetter(0), datau))
x = range(len(y))
plt.errorbar(x, y, u, fmt="kp");
```
