```python
def pdfun(x, y, bins, range, std=False):
    bin_edges   = np.linspace(*range, bins + 1)
    bin_centers = shift_to_bin_centers(bin_edges)
    bin_numbers = np.digitize(x, bin_edges)
    df          = pd.DataFrame(dict(bin=bin_numbers, x=x, y=y))
#    df          = df.sort_values("bin")
    gb          = df.groupby("bin").y
    mean        = gb.mean()
    std_        = gb.std () if std else gb.std() / gb.size()**0.5
    return bin_centers, mean.values, std_.values
```
